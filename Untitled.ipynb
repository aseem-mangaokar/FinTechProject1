```python
alc_consumption_timeseries_csv = Path("Resources/apparent_per_capita_alcohol_consumption_1977_2018.csv.csv")
```

```python
alc_consumption_timeseries_df = pd.read_csv(alc_consumption_timeseries_csv, index_col = 'year', infer_datetime_format = True)
```

```python
alc_consumption_timeseries_df.head()
```

```python
alc_consumption_timeseries_df.dropna()
```

```python
alc_consumption_timeseries_df.drop(columns=['ethanol_beer_gallons_per_capita', 'ethanol_wine_gallons_per_capita', 'ethanol_spirit_gallons_per_capita'], inplace = True)
```

```python
alc_consumption_timeseries_df.drop(columns=['ethanol_all_drinks_gallons_per_capita'], inplace = True)
```

```python
alc_consumption_timeseries_df.head()
```

```python
alc_consumption_timeseries_df.hvplot.line(
    'year',
    'number_of_drinks_total',
    xlabel= 'year',
    ylabel= 'number of drinks',
    groupby= 'state',
    title = "Total Number of Drinks per Capita by State"
)
```

```python
alc_consumption_timeseries_df = alc_consumption_timeseries_df.reset_index()
```

```python
alc_consumption_timeseries_df.dtypes
```

```python
mask = (alc_consumption_timeseries_df['year'] > 2000) & (alc_consumption_timeseries_df['year'] <= 2018)
print(alc_consumption_timeseries_df.loc[mask])
```
